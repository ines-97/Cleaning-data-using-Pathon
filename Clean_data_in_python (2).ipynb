## Etape2: Nettoyer les données

**1.Enlever les valeurs Nan**

```python
df_product=product.dropna(subset=['Order ID'])
df_product
```

**Verification**

```python
df_product.isnull().sum()# nombre des valeurs NaN = 0
```

**2.Enlever les duplication**

```python
df_product.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
# `subset=None` signifie que chaque colonne est utilisée pour déterminer si deux rangées sont différentes ; Pour modifier cela, spécifiez les colonnes sous forme de tableau
# `inplace=True` signifie que la structure des données est modifiée et que les lignes en double ont disparu
#'keep=first' lesser le 1er et enlever le 2eme ligne
```

verification

```python
df_product.duplicated().sum()
```

**Enlever les données erronées**

```python
Errones=df_product[df_product['Order ID'].str[0:2]=='Or']
Errones
```

```python
df_product=df_product[df_product['Order ID'].str[0:2]!='Or']
df_product
```

```python
df_product['Order Date']
```
